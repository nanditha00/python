```python
#labelling the x and y axis
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array(['clad','it','csen','lang','single','several','mech','physics','vedc'])
ypoints = np.array([85,90,96,65,71,96,85,95,79])

plt.plot(xpoints, ypoints)

plt.title("sem1 scores")
plt.xlabel("subjects")
plt.ylabel("scores")

plt.show()
```

```python
#labelling the x and y axis and changing the fonts
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array(['clad','it','csen','lang','single','several','mech','physics','vedc'])
ypoints = np.array([85,90,96,65,71,96,85,95,79])
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(xpoints, ypoints)

plt.title("sem1 scores",fontdict=font1)
plt.xlabel("subjects",fontdict=font2)
plt.ylabel("scores",fontdict=font2)

plt.show()
```
